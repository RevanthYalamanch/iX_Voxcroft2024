```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df = pd.read_csv('cleandata_processed.csv', index_col=0)
df
```

```python
df['Inter Frequency'] = np.zeros
```

```python
name_counts = df['Name'].value_counts()
name_counts
```

```python
df = df.merge(name_counts, on = 'Name', how = 'left')
```

```python
df
```

```python
for num in range(0, len(df['Article_Topic_Keyword_Identified'])):
    df.at[num, 'Article_Topic_Keyword_Identified'] = eval(df.at[num, 'Article_Topic_Keyword_Identified'])
    df.at[num, 'Article_Subject_Keyword_Identified'] = eval(df.at[num, 'Article_Subject_Keyword_Identified'])
```

```python
subject_list = []
for entry in df['Article_Subject_Keyword_Identified']:
        for word in entry:
            if word != 'nan':
                subject_list.append(word)
unique_subjects = list(set(subject_list))
unique_subjects
```

```python
topic_list = []
for entry in df['Article_Topic_Keyword_Identified']:
        for word in entry:
            if word != 'nan':
                topic_list.append(word)
unique_topics = list(set(topic_list))
unique_topics
```

```python
df['Topic Key'] = df['Article_Topic_Keyword_Identified'].apply(lambda x: x.copy())

#for topic in unique_topics:
    #unique_topics.index(topic for topic in topic_list)

# Function to map values
def map_topics(topic_list):
    for topic in topic_list:
        if topic == 'nan':
            return
        else:
            return [unique_topics.index(topic) for topic in topic_list]

# Apply the function to each element in the 'Topic Key' column
df['Topic Key'] = df['Topic Key'].apply(map_topics)
```
