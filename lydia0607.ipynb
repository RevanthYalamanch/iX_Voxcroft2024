```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

```python
people = []
for entry in df["Article_Content_People_AI_Model"]:
    entry = str(entry)
    entry = entry.split("|")
    for word in entry:
        if word not in people:
            people.append(word)

entities = []
for entry in df["Article_Content_Entities_AI_Model"]:
    entry = str(entry)
    entry = entry.split("|")
    for word in entry:
        if word not in entities:
            entities.append(word)

people
entities

```
