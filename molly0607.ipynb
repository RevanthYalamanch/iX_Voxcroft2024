```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import spacy
import seaborn as sns

```

```python
df = pd.read_excel("/Users/mollyrudisill/Voxcroft/20240603_1651_GF SA Safety_Last30Days.xlsx")
```

```python
def collect_all_words(df, columns):
    # Initialize an empty list to store words
    all_words = []
    
    # Iterate over each specified column
    for column in columns:
        # Iterate over each value in the specified column
        for value in df[column]:
            # Check if the value is a string
            if isinstance(value, str):
                # Split the value by "|" and extend the list with the words
                words = value.split("|")
                all_words.extend(words)
    
    return all_words

wordswords = collect_all_words(df, ["Article_Content_People_AI_Model"])
wordswords
```

```python
len(wordswords)
```

```python
def remove_prefix_the(word_list):
    return [word[4:] if word.startswith('The ') else word for word in word_list]
```

```python
def remove_at_sign(word_list):
    return [word[4:] if word.startswith('@') else word for word in word_list]
```

```python
def remove_https(word_list):
     return [word for word in word_list if not word.startswith("https:")]
```

```python
wordswords= remove_prefix_the(wordswords)
wordswords =  remove_at_sign(wordswords)
wordswords = remove_https(wordswords)
wordswords
```

```python
from collections import Counter
entities_count = Counter(wordswords)

for entity, count in entities_count.items():
    print(f'{entity}: {count}')
```

```python
counted = pd.DataFrame(list(entities_count.items()), columns=['Word', 'Count'] )

counted_sorted = counted.sort_values(by="Count", ascending=False)
counted_sorted
```
